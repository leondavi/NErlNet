```python
import csv
with open('mnist_tiles_3_confusion_matrices.csv', 'w', newline='') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=model_performance_dict.keys())
    
    # Writing the header (column names)
    writer.writeheader()
    
    # Writing the rows of data
    for i in range(len(list(model_performance_dict.values))):  # Assume all lists have the same length
        row = {key: model_performance_dict[key][i] for key in model_performance_dict}
        writer.writerow(row)
```
