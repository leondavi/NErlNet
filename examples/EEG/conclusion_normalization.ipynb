```python
import pandas as pd

# Assuming your DataFrames are named model_performence1 and model_performence2
# Merge DataFrames on 'Worker' and 'Class'
merged_df = pd.merge(model_performence1, model_performence2, on=['Worker', 'Class'], suffixes=('_df1', '_df2'))

# Convert columns to numeric in case there are any non-numeric values
numeric_cols = ['TN_df1', 'FP_df1', 'FN_df1', 'TP_df1', 'TN_df2', 'FP_df2', 'FN_df2', 'TP_df2']
merged_df[numeric_cols] = merged_df[numeric_cols].apply(pd.to_numeric)

# Aggregate the relevant columns
merged_df['TN'] = merged_df['TN_df1'] + merged_df['TN_df2']
merged_df['FP'] = merged_df['FP_df1'] + merged_df['FP_df2']
merged_df['FN'] = merged_df['FN_df1'] + merged_df['FN_df2']
merged_df['TP'] = merged_df['TP_df1'] + merged_df['TP_df2']

# Recalculate metrics with division by zero handling
merged_df['Accuracy'] = (merged_df['TP'] + merged_df['TN']) / (merged_df['TP'] + merged_df['TN'] + merged_df['FP'] + merged_df['FN']).replace(0, float('nan'))

merged_df['Precision'] = (merged_df['TP'] / (merged_df['TP'] + merged_df['FP'])).replace([float('inf'), float('nan')], 0)
merged_df['Recall'] = (merged_df['TP'] / (merged_df['TP'] + merged_df['FN'])).replace([float('inf'), float('nan')], 0)
merged_df['True Negative Rate'] = (merged_df['TN'] / (merged_df['TN'] + merged_df['FP'])).replace([float('inf'), float('nan')], 0)

merged_df['Balanced Accuracy'] = (merged_df['Recall'] + merged_df['True Negative Rate']) / 2
merged_df['Informedness'] = merged_df['Recall'] + merged_df['True Negative Rate'] - 1
merged_df['F1'] = (2 * merged_df['Precision'] * merged_df['Recall'] / (merged_df['Precision'] + merged_df['Recall'])).replace([float('inf'), float('nan')], 0)

# Replace NaN with 0 for all metrics
merged_df = merged_df.fillna(0)

# Drop old columns including metrics
merged_df = merged_df.drop(columns=[
    'TN_df1', 'FP_df1', 'FN_df1', 'TP_df1', 'TN_df2', 'FP_df2', 'FN_df2', 'TP_df2',
    'Accuracy_df1', 'Balanced Accuracy_df1', 'Precision_df1', 'Recall_df1', 'True Negative Rate_df1', 'Informedness_df1', 'F1_df1',
    'Accuracy_df2', 'Balanced Accuracy_df2', 'Precision_df2', 'Recall_df2', 'True Negative Rate_df2', 'Informedness_df2', 'F1_df2',
    'Unnamed: 0_df1', 'Unnamed: 0_df2'
])

# Define the desired column order
desired_order = [
    'Worker', 'Class', 'TN', 'FP', 'FN', 'TP', 'Accuracy', 'Balanced Accuracy', 
    'Precision', 'Recall', 'True Negative Rate', 'Informedness', 'F1'
]

# Reorder columns
merged_df = merged_df[desired_order]

# Display the final merged DataFrame
merged_df
```

```python
plt.figure(figsize=(12,5))
sns.barplot(data=merged_df.sort_values('Worker') , x='Worker' , y='F1',hue='Class')
plt.title("Worker - F1 Score")
plt.legend(title='Class', bbox_to_anchor=(1.15, 1), loc='upper right') 
plt.show()
plt.figure(figsize=(12,5))
sns.barplot(data=merged_df.sort_values('Worker') , x='Worker' , y='Accuracy',hue='Class')
plt.title("Worker - Accuracy")
plt.legend(title='Class', bbox_to_anchor=(1.15, 1), loc='upper right') 
plt.show()
plt.figure(figsize=(12,5))
sns.barplot(data=merged_df.sort_values('Worker') , x='Worker' , y='Precision',hue='Class')
plt.title("Worker - Precision")
plt.legend(title='Class', bbox_to_anchor=(1.15, 1), loc='upper right') 
plt.show()
plt.figure(figsize=(12,5))
sns.barplot(data=merged_df.sort_values('Worker') , x='Worker' , y='Recall',hue='Class')
plt.title("Worker - Recall")
plt.legend(title='Class', bbox_to_anchor=(1.15, 1), loc='upper right') 
plt.show()
```
