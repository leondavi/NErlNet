```python
import pandas as pd

# Assuming your DataFrames are named model_performence1 and model_performence2
# Merge DataFrames on 'Worker' and 'Class'
merged_df = pd.merge(model_performence1, model_performence2, on=['Worker', 'Class'], suffixes=('_df1', '_df2'))

# Aggregate the relevant columns
merged_df['TN'] = merged_df['TN_df1'] + merged_df['TN_df2']
merged_df['FP'] = merged_df['FP_df1'] + merged_df['FP_df2']
merged_df['FN'] = merged_df['FN_df1'] + merged_df['FN_df2']
merged_df['TP'] = merged_df['TP_df1'] + merged_df['TP_df2']

# Recalculate metrics
merged_df['Accuracy'] = (merged_df['TP'] + merged_df['TN']) / (merged_df['TP'] + merged_df['TN'] + merged_df['FP'] + merged_df['FN'])
merged_df['Precision'] = merged_df['TP'] / (merged_df['TP'] + merged_df['FP'])
merged_df['Recall'] = merged_df['TP'] / (merged_df['TP'] + merged_df['FN'])
merged_df['True Negative Rate'] = merged_df['TN'] / (merged_df['TN'] + merged_df['FP'])
merged_df['Balanced Accuracy'] = (merged_df['Recall'] + merged_df['True Negative Rate']) / 2
merged_df['Informedness'] = merged_df['Recall'] + merged_df['True Negative Rate'] - 1
merged_df['F1'] = 2 * (merged_df['Precision'] * merged_df['Recall']) / (merged_df['Precision'] + merged_df['Recall'])

# Replace NaN with 0
merged_df = merged_df.fillna(0)

# Drop old columns including metrics
merged_df = merged_df.drop(columns=[
    'TN_df1', 'FP_df1', 'FN_df1', 'TP_df1', 'TN_df2', 'FP_df2', 'FN_df2', 'TP_df2',
    'Accuracy_df1', 'Balanced Accuracy_df1', 'Precision_df1', 'Recall_df1', 'True Negative Rate_df1', 'Informedness_df1', 'F1_df1',
    'Accuracy_df2', 'Balanced Accuracy_df2', 'Precision_df2', 'Recall_df2', 'True Negative Rate_df2', 'Informedness_df2', 'F1_df2',
    'Unnamed: 0_df1', 'Unnamed: 0_df2'	
])

# Define the desired column order
desired_order = [
    'Worker', 'Class', 'TN', 'FP', 'FN', 'TP', 'Accuracy', 'Balanced Accuracy', 
    'Precision', 'Recall', 'True Negative Rate', 'Informedness', 'F1'
]

# Reorder columns
merged_df = merged_df[desired_order]

merged_df
```
