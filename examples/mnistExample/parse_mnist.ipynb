```python
dfFull = pd.read_csv(CSV_PATH)
dfFull["label"] = dfFull["label"].astype("category")
label_one_hot = pd.get_dummies(dfFull["label"], prefix="label").astype(int)
df = pd.concat([dfFull, label_one_hot], axis=1)
```
